# **Data Visualization**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample: Fuel Type distribution
df = pd.read_csv('used_car_price_dataset.csv')  # Replace with your file path
fuel_counts = df['fuel_type'].value_counts()

plt.figure(figsize=(8,5))
bars = plt.bar(fuel_counts.index, fuel_counts.values, color='skyblue')
plt.title("Fuel Type Distribution")
plt.xlabel("Fuel Type")
plt.ylabel("Number of Cars")

# Add data labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10, int(yval), ha='center')

plt.tight_layout()
plt.show()
```

```python
# --- Pie Chart: Fuel Type Distribution ---
plt.figure(figsize=(6, 6))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Fuel Type Distribution - Pie Chart")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# --- Line Chart: Fuel Type Trend (Dummy Time Series) with Data Labels ---
# Assuming dummy time-series data for visualization purpose
# If real time data is available, modify accordingly
# Simulate sample monthly fuel-type usage
dummy_data = pd.DataFrame({
    'Month': pd.date_range(start='2024-01-01', periods=6, freq='M'),
    'Petrol': [150, 160, 170, 180, 175, 190],
    'Diesel': [80, 85, 82, 90, 88, 95],
    'Electric': [20, 22, 24, 25, 28, 30]
})

plt.figure(figsize=(10, 6))
for col in dummy_data.columns[1:]:
    plt.plot(dummy_data['Month'], dummy_data[col], marker='o', label=col)
    for i, val in enumerate(dummy_data[col]):
        plt.text(dummy_data['Month'][i], val + 2, str(val), ha='center', fontsize=9)
plt.title("Fuel Type Trend Over Time - Line Chart")
plt.xlabel("Month")
plt.ylabel("Number of Cars")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- Box Plot: Mileage Distribution by Fuel Type with Data Labels ---
plt.figure(figsize=(10, 6))
sns.boxplot(x='fuel_type', y='mileage_kmpl', data=df, palette='Set2')
plt.title("Mileage Distribution by Fuel Type - Box Plot")
plt.xlabel("Fuel Type")
plt.ylabel("Mileage (km/l)")

# Add data labels: show median values
medians = df.groupby('fuel_type')['mileage_kmpl'].median().values
fuel_types = df['fuel_type'].unique()
for i, median in enumerate(medians):
    plt.text(i, median + 0.5, f"Median: {median:.1f}", ha='center', color='black', fontsize=9)

plt.tight_layout()
plt.show()
```
